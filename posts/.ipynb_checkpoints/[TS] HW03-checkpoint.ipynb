다음의 모형들에 의해 설명되는 확률과정 $\{Z_t\}$는 정상성을 갖는가? 단 $ε_t ∼ W N(0, 1)$.

`-` 정상성 만족하기 위해서는

$E(Z_t) = \mu$

$Var(Z_t) = \sigma^2$

$Cov(Z_t,Z_{t+h}) = \gamma_h$

## (1)

$Z_t = ε_t − ε_{t−1} − ε_{t−2}$

$E(Z_t) = E(ε_t − ε_{t−1} − ε_{t−2}) = E(ε_t)-E(ε_{t-1}) - E(ε_{t-2}) = 0$

$Var(Z_t) = Var(ε_t − ε_{t−1} − ε_{t−2}) = Var(ε_t)+Var(\epsilon_{t-1}) + Var(ε_{t−2}) = 1+1+1=3$

- $\epsilon_t$는 WN이므로 $Cov(\epsilon_t, \epsilon_{t+h})=0$

$Cov(Z_t,Z_{t-h}) = Cov(ε_t - ε_{t-1} - ε_{t-2}, ε_{t-h} - ε_{t-h-1} - ε_{t-h-2})$

- $\epsilon$은 WN이므로 각각 COV취해줄 수 있꼬,, 상수값 나오니까 정상 시계열 조건을 만족한다.

주어진 시계열 모형이 정상성을 만족하는지 확인하기 위해 평균과 공분산을 확인할 것입니다.

먼저 주어진 시계열 모형은 다음과 같습니다:

\[ Z_t = \epsilon_t - \epsilon_{t-1} - \epsilon_{t-2} \]

여기서 \( \epsilon_t \)는 백색소음(White Noise)이라 가정하고 있습니다.

1. **평균 확인:**
\[ E(Z_t) = E(\epsilon_t - \epsilon_{t-1} - \epsilon_{t-2}) \]

먼저 \(E(\epsilon_t) = 0\)이고, 독립성으로 인해 \(E(\epsilon_{t-1}) = E(\epsilon_{t-2}) = 0\)이므로

\[ E(Z_t) = 0 - 0 - 0 = 0 \]

평균이 0이므로 평균 조건은 만족합니다.

2. **분산 확인:**
\[ Var(Z_t) = Var(\epsilon_t - \epsilon_{t-1} - \epsilon_{t-2}) \]

여기서 각 \( \epsilon_t \)는 서로 독립이므로,

\[ Var(Z_t) = Var(\epsilon_t) + Var(\epsilon_{t-1}) + Var(\epsilon_{t-2}) \]

주어진 정보에 따르면 \(Var(\epsilon_t) = 1\)이므로,

\[ Var(Z_t) = 1 + 1 + 1 = 3 \]

분산이 상수인 경우이므로 분산 조건도 만족합니다.

3. **공분산 확인:**
\[ Cov(Z_t, Z_{t-h}) = Cov(\epsilon_t - \epsilon_{t-1} - \epsilon_{t-2}, \epsilon_{t-h} - \epsilon_{t-h-1} - \epsilon_{t-h-2}) \]

이때, 각 \( \epsilon_t \)는 서로 독립이므로 \(Cov(\epsilon_t, \epsilon_{t-h}) = 0\)입니다.

따라서 \(Cov(Z_t, Z_{t-h}) = 0\)이며, 공분산이 시간에 의존하지 않습니다.

결론적으로, 주어진 시계열 모형은 평균과 분산이 상수이며, 공분산이 시간에 독립적이므로 정상성을 만족합니다.

이제 새로운 모형인 \(Z_t = \epsilon_t \epsilon_{t-1} + \epsilon_{t-2}\)에 대한 분석을 해보겠습니다.

## (2)

$Z_t = ε_tε_{t−1} + ε_{t−2}$

-> 정상시계열

$E(Z_t) = 0$

$Var(Z_t) = V(\epsilon_t \epsilon_{t-1} + \epsilon_{t-2}) = Var(\epsilon_{t-1} \epsilon_{t-2} + \epsilon_{t-3}) = V(Z_{t-1})$

$Var(Z_t) = Var(Z_{t-1}) = Var(Z_{t-2})= \dots$

$cov(Z_t, Z_{t-1}) = cov(Z_{t-1}, Z_{t-2})$

새로운 시계열 모형이 정상성을 만족하는지 확인하기 위해 다음을 살펴보겠습니다.

주어진 모형은 다음과 같습니다:
\[ Z_t = \epsilon_t \epsilon_{t-1} + \epsilon_{t-2} \]

여기서 \( \epsilon_t \)은 백색소음(White Noise)이라 가정하고 있습니다.

1. **평균 확인:**
\[ E(Z_t) = E(\epsilon_t \epsilon_{t-1} + \epsilon_{t-2}) \]

\(E(\epsilon_t) = 0\)이므로
\[ E(Z_t) = E(\epsilon_t \epsilon_{t-1}) + E(\epsilon_{t-2}) \]

두 항 모두 독립성으로 인해 0이므로,
\[ E(Z_t) = 0 \]

평균이 0이므로 평균 조건은 만족합니다.

2. **분산 확인:**
\[ Var(Z_t) = Var(\epsilon_t \epsilon_{t-1} + \epsilon_{t-2}) \]

각 \( \epsilon_t \)는 서로 독립이 아닙니다. 하지만 \( \epsilon_t \epsilon_{t-1} \)의 경우에는 곱으로 인해 공분산이 0이 됩니다. 따라서,

\[ Var(Z_t) = Var(\epsilon_{t-2}) \]

주어진 정보에 따르면 \(Var(\epsilon_{t-2}) = 1\)이므로,
\[ Var(Z_t) = 1 \]

분산이 상수인 경우이므로 분산 조건도 만족합니다.

3. **공분산 확인:**
\[ Cov(Z_t, Z_{t-h}) = Cov(\epsilon_t \epsilon_{t-1} + \epsilon_{t-2}, \epsilon_{t-h} \epsilon_{t-h-1} + \epsilon_{t-h-2}) \]

\( \epsilon_t \)와 \( \epsilon_{t-h} \)는 서로 독립이 아닙니다. 하지만 \( \epsilon_t \epsilon_{t-1} \)와 \( \epsilon_{t-h} \epsilon_{t-h-1} \)의 경우에는 각각 독립이므로,

\[ Cov(Z_t, Z_{t-h}) = Cov(\epsilon_{t-2}, \epsilon_{t-h-2}) \]

따라서, \(Cov(Z_t, Z_{t-h})\)는 상수이며 시간에 독립적입니다.

결론적으로, 주어진 새로운 시계열 모형도 평균과 분산이 상수이고, 공분산이 시간에 독립적이므로 정상성을 만족합니다.

## (3)

$Z_t = A sin \left( \dfrac{2}{3} πt + U \right)$, 단 A는 평균이 0이고 분산이 1인 확률변수이고, $U$ 는 상수이다.

```r
# 필요한 라이브러리 불러오기
library(ggplot2)

# 시뮬레이션을 위한 난수 시드 설정
set.seed(123)

# 시뮬레이션 횟수
n <- 100

# A는 평균이 0이고 분산이 1인 확률변수
A <- rnorm(n, mean = 0, sd = 1)

# U는 상수
U <- 2

# 시뮬레이션된 데이터 생성
t <- seq(1, n, by = 1)
Z <- A * sin((2/3) * pi * t + U)

# 생성된 데이터 시각화
ggplot(data.frame(t, Z), aes(x = t, y = Z)) +
  geom_line() +
  labs(title = "Simulation of Z_t", x = "Time (t)", y = "Z_t")
```

제시된 시계열 모형인 \[Z_t = A \sin\left(\frac{2}{3}\pi t + U\right)\]에 대해서 정상성을 검토해보겠습니다.

주어진 모형에서 \(A\)는 평균이 0이고 분산이 1인 확률변수이고, \(U\)는 상수입니다.

1. **평균 확인:**
\[E(Z_t) = E(A \sin\left(\frac{2}{3}\pi t + U\right))\]

확률변수 \(A\)의 평균은 0이므로,
\[E(Z_t) = 0\]

평균이 0이므로 평균 조건은 만족합니다.

2. **분산 확인:**
\[Var(Z_t) = Var(A \sin\left(\frac{2}{3}\pi t + U\right))\]

확률변수 \(A\)의 분산이 1이므로,
\[Var(Z_t) = Var(\sin\left(\frac{2}{3}\pi t + U\right))\]

사인 함수의 범위는 -1에서 1이므로 분산은 1을 초과하지 않습니다.

\[Var(Z_t) \leq 1\]

분산이 상수인 경우이므로 분산 조건도 만족합니다.

3. **공분산 확인:**
\[Cov(Z_t, Z_{t-h}) = Cov(A \sin\left(\frac{2}{3}\pi t + U\right), A \sin\left(\frac{2}{3}\pi (t-h) + U\right))\]

두 사인 함수는 서로 독립적이므로 공분산은 0입니다.

\[Cov(Z_t, Z_{t-h}) = 0\]

결론적으로, 주어진 시계열 모형도 평균과 분산이 상수이고, 공분산이 시간에 독립적이므로 정상성을 만족합니다.

## (4)

$Z_t = A sin (πt + U)$, 단 $A$는 평균이 0이고 분산이 1인 확률변수이고, $U$ 는 $[−π, π]$에서의
균등분포(uniform distribution)을 따르는 확률변수이다. $A, U$ 는 서로 독립.

```r
# 필요한 라이브러리 불러오기
library(ggplot2)

# 시뮬레이션을 위한 난수 시드 설정
set.seed(123)

# 시뮬레이션 횟수
n <- 100

# A는 평균이 0이고 분산이 1인 확률변수
A <- rnorm(n, mean = 0, sd = 1)

# U는 [-π, π]에서 균등분포를 따르는 확률변수
U <- runif(n, min = -pi, max = pi)

# 시뮬레이션된 데이터 생성
t <- seq(1, n, by = 1)
Z <- A * sin(pi * t + U)

```

주어진 모형은 다음과 같습니다:

\[ Z_t = A \sin(\pi t + U) \]

여기서:
- \( A \)는 평균이 0이고 분산이 1인 확률변수입니다.
- \( U \)는 \([- \pi, \pi]\)에서 균등분포를 따르는 확률변수입니다.
- \( A \)와 \( U \)는 서로 독립입니다.

이 모형이 정상성을 만족하는지 확인하기 위해 다음을 살펴보겠습니다:

1. **평균 확인:**
\[ E(Z_t) = E(A \sin(\pi t + U)) \]

기댓값 \( E(A) \)가 0이므로
\[ E(Z_t) = 0 \]

평균이 0이므로 평균 조건은 만족합니다.

2. **분산 확인:**
\[ Var(Z_t) = Var(A \sin(\pi t + U)) \]

확률변수 \( A \)의 분산이 1이므로
\[ Var(Z_t) = Var(\sin(\pi t + U)) \]

여기서 \( \sin(\pi t + U) \)의 값은 항상 -1에서 1 사이의 값을 가지므로 분산은 1을 초과하지 않습니다.

\[ Var(Z_t) \leq 1 \]

분산이 상수인 경우이므로 분산 조건도 만족합니다.

3. **공분산 확인:**
\[ Cov(Z_t, Z_{t-h}) = Cov(A \sin(\pi t + U), A \sin(\pi (t-h) + U)) \]

확률변수 \( A \)와 \( U \)가 서로 독립이므로
\[ Cov(Z_t, Z_{t-h}) = Cov(\sin(\pi t + U), \sin(\pi (t-h) + U)) \]

이때, 사인 함수의 주기성 특성을 고려하면 두 항 간의 공분산이 상당 부분 0으로 갈 것입니다. 하지만 정확한 값은 주어진 특정한 형태에 따라 달라집니다.

따라서, 주어진 모형도 평균과 분산이 상수이고, 공분산이 시간에 독립적이므로 정상성을 만족합니다.

확률변수 \( Z_t \)의 공분산을 구하려면 다음과 같이 공분산의 정의를 사용합니다:

\[ \text{Cov}(Z_t, Z_{t-h}) = \text{Cov}(A \sin(\pi t + U), A \sin(\pi (t-h) + U)) \]

여기서 \( A \)와 \( U \)는 독립이므로, 공분산을 계산할 때는 다음과 같이 활용할 수 있습니다:

\[ \text{Cov}(Z_t, Z_{t-h}) = \text{Cov}(\sin(\pi t + U), \sin(\pi (t-h) + U)) \]

이제 사인 함수의 주기성 특성을 이해해야 합니다. 사인 함수는 주기적인 함수로, 주기는 \(2\pi\)입니다. 따라서, \( \sin(\pi t + U) \)와 \( \sin(\pi (t-h) + U) \)는 같은 주기를 갖습니다.

\[ \sin(\pi t + U) \]의 경우, \( U \)는 \([- \pi, \pi]\)에서 균등분포를 따르므로, 이 사인 함수의 값은 \([-1, 1]\) 범위에서 주기적으로 변할 것입니다.

이제 두 사인 함수 간의 공분산을 생각해봅시다. 두 사인 함수는 주기적이므로, 이 둘 간의 공분산은 사인 함수의 주기와 관련이 있습니다. 사인 함수의 주기는 \(2\pi\)이므로, 주기에 따라 두 사인 함수는 비슷한 패턴을 갖게 됩니다.

이때, 만약 \(U\)가 작은 값(절대적으로 작은 각도 변화)을 갖는다면, 두 사인 함수의 주기적인 패턴이 비슷해지고 공분산이 상당히 0에 가까워집니다. 왜냐하면 주기적인 함수는 주기가 다를 때 공분산이 작아지기 때문입니다.

그러나 이 모든 것은 구체적인 값을 가정해야 정확한 값을 계산할 수 있다는 것을 의미합니다. 이는 특정한 값에 대한 수치적 계산이 필요하며, 일반적인 형태에서는 단순한 분석으로는 정확한 값을 도출하기 어렵습니다.

Z_{t}=\epsilon_t + 1  &t: \text{홀수}
\end{cases}$

```r
Cov을 보바
```

```r
\[ \text{Cov}(Z_t, Z_{t-h}) = \begin{cases}
\text{Cov}(\epsilon_t, \epsilon_{t-h}) & \text{if } t \text{ is even} \\
\text{Cov}(\epsilon_t + 1, \epsilon_{t-h} + 1) & \text{if } t \text{ is odd}
\end{cases} \]
```

```r
# 필요한 라이브러리 불러오기
library(ggplot2)

# 홀수와 짝수 시점에 대한 시뮬레이션 횟수
n <- 100

# 백색소음 생성 함수
generate_white_noise <- function(n) {
  rnorm(n)
}

# 시뮬레이션 데이터 생성
epsilon_even <- generate_white_noise(n)
epsilon_odd <- generate_white_noise(n)

# 홀수와 짝수 시점에 대한 Z_t 계산
Z_even <- epsilon_even
Z_odd <- epsilon_odd + 1

# 데이터 프레임 생성
data <- data.frame(
  time = rep(1:n, each = 2),
  Z = c(Z_even, Z_odd)
)

# 생성된 데이터 시각화
ggplot(data, aes(x = time, y = Z)) +
  geom_line() +
  labs(title = "Simulation of Z_t", x = "Time", y = "Z_t") +
  theme_minimal()
```

```r
set.seed(1)
cov(Z_even,Z_odd)
```

```r
set.seed(1)
epsilon_even <- generate_white_noise(n) + 3
epsilon_odd <- generate_white_noise(n) + 3

# 홀수와 짝수 시점에 대한 Z_t 계산
Z_even <- epsilon_even
Z_odd <- epsilon_odd + 1
cov(Z_even,Z_odd)
```

$ε_t$가 $WN(0, σ^2)$를 따를 때, 다음과 같은 확률과정 모형에 대하여 각 물음에 답하여라

$$Z_t − 0.8Z_{t−1} = ε_t$$

## (1)

모형을 $ϕ(B)(Z_t − µ) = θ(B)ε_t$로 표현하고, $ϕ(B), θ(B)$ 그리고 $µ$를 명시하여라.


$(1-0.8B)Z_t = \epsilon_t$

$\mu-0.8\mu = 0 \rightarrow \mu = 0$

$ϕ(B)=1-0.8B$

$\theta(B) = 1$

$\mu=0$

## (2)

모형은 AR(p), MA(q) 혹은 ARMA(p, q)모형 중 어느 것인가? p와 q도 함께 명시하여라.

AR(1)

## (3)

ACF $ρ_k, k = 1, . . . , 5$를 계산하여라.

$lag=0, Cov(Z_t,Z_t)$ 를 구하자.

$\gamma_0 = [1+0.8^2 +0.8^4 +\dots ] \sigma^2=\dfrac{\sigma^2}{1-0.8^2}$

$lag=1$

| | $\epsilon_t$ | $\epsilon_{t-1}$ | $\epsilon_{t-2}$ | $\dots$|
|--|--|--|--|--|
|$Z_t$|1|0.8|$0.8^2$|$\dots$|
|$Z_{t-1}$|0|1|0.8|$\dots$|

$\gamma_1 = [0+0.8 + 0.8^3 + \dots] \sigma^2 = \dfrac{0.8 \sigma^2}{1-0.8^2}$

즉

$\gamma_0 = \dfrac{\sigma^2}{1-0.8^2}$

$\gamma_1 = \dfrac{0.8\sigma^2}{1-0.8^2}$

$\gamma_2 = \dfrac{0.8^2\sigma^2}{1-0.8^2}$

$\gamma_3 = \dfrac{0.8^3\sigma^2}{1-0.8^2}$

$\gamma_4 = \dfrac{0.8^4\sigma^2}{1-0.8^2}$

$\gamma_5 = \dfrac{0.8^5\sigma^2}{1-0.8^2}$


$\rho_k = \dfrac{\gamma_k}{\gamma_0}$

$\rho_1 = \dfrac{\gamma_1}{\gamma_0}=0.8$

$\rho_2 = \dfrac{\gamma_2}{\gamma_0}=0.8^2$

$\rho_3 = \dfrac{\gamma_3}{\gamma_0}=0.8^3$

$\rho_4 = \dfrac{\gamma_4}{\gamma_0}=0.8^4$

$\rho_5 = \dfrac{\gamma_5}{\gamma_0}=0.8^5$

## (4)

PACF $ϕ_{kk}, k = 1, . . . , 5$를 계산하여라.


$ϕ_{11}=\rho_1 = 0.8$

AR(1)모형이므로 $k \geq 2$이면 $ϕ_{kk}=0$이다.

## (5) 다시...

위에서 구한 $ρ_k, ϕ_{kk}$ 의 상관도표를 그려라

```r
sim_ar <- function(n, mu, phi){
 ### n : sample size
 ### mu : mean
 ### phi : p-dim coefficients

 p <- length(phi)
 z <- rnorm(n+100) #epsilon ~ WN(0, sigma^2), iid N(0,1)
 delta <- (1-sum(phi))*mu

 for (k in (length(phi)+1):(n+100)){

 z[k] <- delta + sum(z[(k-1):(k-p)]*phi) + rnorm(1) 
 }

 return(z[-(1:100)])
}
```

```r
z <- sim_ar(100, 0, phi=c(0.8)) ##AR(1)
forecast::tsdisplay(z)
```

```r
##AR(1) phi=0.5
z <- arima.sim(n=100, ##order=c(p,d,q) ARMA : d=0, AR : d=q=0
 list(order=c(1,0,0), ar= 0.8), #ar=ϕ1
 rand.gen = rnorm,
 sd = sqrt(1)) #분산
forecast::tsdisplay(z, lag.max=24)
```
